## User statistics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi'] = 200
```

```python
# Load experiment results
experiment_results = pd.read_csv('data/experiment_results_raw.csv')
```

### Filtering responses

```python
# Run the orthographic_equivalencies.py script
%run 'clean_user_responses.py'
```

```python
# Step 1: Get all unique user IDs and emails
unique_emails = experiment_results['prolific_id_email'].dropna().unique()
# Step 2: Remove test users by email addresses
filtered_results  = experiment_results[~experiment_results['prolific_id_email'].isin(excluded_emails) & ~experiment_results['prolific_id_email'].isna()]
```

### Clean out too short and too long responses

```python
responses_list = list(experiment_results['user_free_translation'])
cleaned_responses = ['nan' if type(s) == float or len(str(s.strip())) < 3  or len(str(s.strip())) > 30 else s.lower().strip() for s in responses_list]

experiment_results['user_free_translation'] = cleaned_responses
```

```python
experiment_results.head()
```

### Filter out users

```python
# Excluding test emails
filtered_experiment_results = experiment_results[~experiment_results['prolific_id_email'].isin(excluded_emails)]
```

```python
filtered_experiment_results.head()
```

### Correct names for other_langs

```python
other_langs_dict = {'polish': 'pl', 'польский': 'pl', 'украинский': 'uk'}
filtered_experiment_results = filtered_experiment_results.copy()
other_langs_corrected = [str(x).strip().lower() for x in filtered_experiment_results['other_langs'].to_list()]
for x in range(len(other_langs_corrected)):
    nope = True
    for l in other_langs_dict:
        if l in other_langs_corrected[x]:
            other_langs_corrected[x] = other_langs_dict[l]
            nope = False
    if nope == True:
        other_langs_corrected[x] = 'nan'
filtered_experiment_results['other_langs'] = other_langs_corrected

# Exclude rows where age is less than 18
filtered_experiment_results = filtered_experiment_results[filtered_experiment_results['age'] >= 18]
```

### If person knows the language, exclude theirs answers for that language

```python
# Number of rows before filtering
rows_before_filtering = len(filtered_experiment_results)

# Apply filtering condition
filtered_experiment_results = filtered_experiment_results[
    (filtered_experiment_results['source_language'] != filtered_experiment_results['first_lang']) &
    (~filtered_experiment_results.apply(lambda row: row['source_language'] in str(row['other_langs']), axis=1))
]

# Number of rows after filtering
rows_after_filtering = len(filtered_experiment_results)

# Number of rows excluded
rows_excluded = rows_before_filtering - rows_after_filtering

print(f"Number of participants excluded: {rows_excluded}")
```

```python
# Calculating the desired statistics
num_unique_users = filtered_experiment_results['user_id'].nunique()
avg_age = filtered_experiment_results['age'].mean()
min_age = filtered_experiment_results['age'].min()
max_age = filtered_experiment_results['age'].max()
# Removing duplicate user IDs and getting gender counts
unique_users = filtered_experiment_results.drop_duplicates(subset='user_id')
gender_counts = unique_users['gender'].value_counts()

# Displaying the results
print(f"Number of unique user IDs: {num_unique_users}")
print(f"Average age: {avg_age:.2f}")
print(f"Lowest age: {min_age}")
print(f"Highest age: {max_age}")
print("\nGender counts:")
print(gender_counts)
```

```python

# Calculating the number of unique users per stimulus and language
unique_users_per_stimulus_and_language = filtered_experiment_results.groupby(['source_text_to_be_translated'])['user_id'].nunique()

# Displaying the results
print(unique_users_per_stimulus_and_language)

# Finding the lowest and highest number of unique users
min_unique_users = unique_users_per_stimulus_and_language.min()
max_unique_users = unique_users_per_stimulus_and_language.max()

# Calculating the average number of unique users per stimulus
avg_unique_users_per_stimulus = filtered_experiment_results.groupby('source_text_to_be_translated')['user_id'].nunique().mean()

# Displaying the results
print(f"Lowest number of unique users per stimulus and language: {min_unique_users}")
print(f"Highest number of unique users per stimulus and language: {max_unique_users}")
print(f"Average number of unique users per stimulus: {avg_unique_users_per_stimulus:.2f}")
```

```python
# Calculating the number of unique users per stimulus and language
unique_users_per_stimulus_and_language = filtered_experiment_results.groupby(['source_language'])['user_id'].nunique()

# Displaying the results
print(unique_users_per_stimulus_and_language)

# Finding the lowest and highest number of unique users
min_unique_users = unique_users_per_stimulus_and_language.min()
max_unique_users = unique_users_per_stimulus_and_language.max()

# Calculating the average number of unique users per stimulus
avg_unique_users_per_stimulus = filtered_experiment_results.groupby('source_language')['user_id'].nunique().mean()

# Displaying the results
print(f"Lowest number of unique users per stimulus and language: {min_unique_users}")
print(f"Highest number of unique users per stimulus and language: {max_unique_users}")
print(f"Average number of unique users per stimulus: {avg_unique_users_per_stimulus:.2f}")
```

```python
pd.option_context('mode.use_inf_as_na', True)
# Identify unique users
unique_users = filtered_experiment_results.drop_duplicates(subset='user_id')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn palette to Set2
sns.set_palette("Set2")

# Create subplots with a smaller size
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Plot 1: Bar chart for first language distribution (Seaborn)
sns.countplot(data=unique_users, x='first_lang', ax=axs[0, 0])
axs[0, 0].set_title('First Language Distribution')
axs[0, 0].set_xlabel('First Language')
axs[0, 0].set_ylabel('Count')

# Plot 2: Bar chart for other languages distribution (Seaborn)
other_langs = unique_users['other_langs'].str.split(', ', expand=True).stack()
sns.countplot(data=other_langs.reset_index(level=1), x=0, ax=axs[0, 1])
axs[0, 1].set_title('Other Languages Distribution')
#axs[0, 1].set_xticklabels(['-', 'Polish', 'Ukrainian'])
axs[0, 1].set_xlabel('Other Languages')
axs[0, 1].set_ylabel('Count')

# Plot 3: Pie chart for gender distribution (Matplotlib)
gender_counts = unique_users['gender'].value_counts()
axs[1, 0].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
axs[1, 0].set_title('Gender Distribution')

# Plot 4: Histogram for age distribution (Matplotlib)
axs[1, 1].hist(unique_users['age'], bins=10)
axs[1, 1].set_title('Age Distribution')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Count')

# Set Matplotlib color cycle to match Seaborn palette
for ax in axs.ravel():
    ax.set_prop_cycle(plt.cycler('color', plt.cm.Set2.colors))

# Adjust layout
plt.tight_layout()

# Save the plot in the "data/results/" directory
plt.savefig('data/results/user_statistics_plot.png')

# Show the plots
plt.show()
```
